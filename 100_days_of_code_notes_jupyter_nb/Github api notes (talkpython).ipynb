```python
Repo = namedtuple('Repo', 'name, stars, forks')

def get_github_repos(name, n):
    repos = []
    gh = Github()
    nu = gh.get_user(name)
    
    for repo in nu.get_repos():
        if repo.fork:
            continue
            
        repos.append(Repo(name=repo.name, stars=repo.stargazers_count, forks=repo.forks))
    
    return repos
```

```python
for repo in get_github_repos('vipinreyo', 5):
    print(f'{repo.name}, {repo.stars}, {repo.forks}')
```

```python
for repo in get_github_repos('pybites', 5):
    print(f'{repo.name}, {repo.stars}, {repo.forks}')
```
